## Beetje spelen met Pandas
Let erop dat het databestand in dezelfde map staat als je notebook, anders moet je het volledige pad aangeven.
We lezen een excelbestand, doen wat transformaties, en schrijven het weer weg naar een ander bestand.

```python
mijndata = pd.read_excel("Alle_4599_chunks_EN_news.xlsx", header=None)
```

```python
mijndata = mijndata.drop(0, axis=1)
```

```python
mijndata.columns=['onzetext']
```

```python
mijndata['kijkeens'] = mijndata['onzetext'].str.upper()
```
